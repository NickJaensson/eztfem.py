```python
import eztfem as ezt
import gmsh

gmsh.initialize()
gmsh.option.setNumber("Mesh.CharacteristicLengthMin", 0.01)
gmsh.option.setNumber("Mesh.CharacteristicLengthMax", 0.04)
```

```python
import unit_square

mesh = ezt.gmsh_mesh2d(element_type='tri3', model_builder=unit_square.main)
# ezt.plot_points_curves(mesh)
mesh_pv = ezt.generate_pyvista_mesh(mesh)
ezt.plot_mesh_pv(mesh_pv, show_edges=True, window_size=(800, 400))
```
