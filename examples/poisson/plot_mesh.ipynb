```python
import functools
import eztfem as ezt
from unit_square import build_unit_square
```

```python
mesh = ezt.gmsh_mesh2d(
    model_builder=functools.partial(build_unit_square, 
                                    element_size=[0.1, 0.1, 0.04, 0.04]),
    gmsh_options={
        "Mesh.ElementOrder": 2,
        "Mesh.HighOrderOptimize": 1,
    },
)
```

```python
ezt.plot_points_curves(mesh)
```

```python
mesh_pv = ezt.generate_pyvista_mesh(mesh)
```

```python
ezt.plot_mesh_pv(mesh_pv, show_edges=True, window_size=(800, 400))
```
